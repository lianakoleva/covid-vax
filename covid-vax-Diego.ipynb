```python
import folium
from folium import plugins
import pandas as pd
```

```python
# pd.set_option('display.max_rows', None)
df = pd.read_csv("C:/Users/dp846/Downloads/vaccination-by-zip-code.csv")
df = df.drop(['1st Dose - Daily','1st Dose - Cumulative','1st Dose - Percent Population','Vaccine Series Completed - Daily','Vaccine Series Completed - Cumulative','Vaccine Series Completed  - Percent Population','Population','Row_ID'], axis = 1)
df = df.dropna()
```

```python
coord_list = [str(i).replace('POINT','').strip().replace('(','').replace(')','').split() for i in df['ZIP Code Location']]

long_coord = [float(i[0]) for i in coord_list]

lat_coord = [float(i[1]) for i in coord_list]

df['Latitude'] = lat_coord

df['Longitude'] = long_coord

df
```

```python
loc = df[['Latitude', 'Longitude']]
```

```python
m = folium.Map([41.8781, -87.6298], zoom_start=11)
m.add_child(plugins.HeatMap(loc, radius=15))
m
```
